# Let's work with numPy!
Import numpy as follows:

```python
import numpy as np #np here is an alias/nickname, could have been anything, it saves time typing
```

```python
# numpy has a function called random
# if you need to call a function from numpy you can do so as follows:
# optional:
#np.random.seed(42) # freezes the randomizer and makes results reproducable
scores = np.random.normal(75, 10, 1000) # mean=75, sd=10, n=1000
print(scores)
```

```python
type(scores)
```

```python
# numpy arrays are similar to lists
```

```python
scores*5 # works because scores object is a numpy array
```

```python
mylist = [1,2,3]
```

```python
type(mylist)
```

```python
mylist*5 # the output here is different than with a numpy array
```

```python
# what if you want to round numbers in a numpy array?
np.round(scores,2) # if you want to round to two decimal points
```

```python
# how about choosing elements inside an array?
# how do I select the first item in scores?
scores[0] # returns the first element in scores
```

```python
scores[1] # returns the second etc.
```

```python
# what if you want to return scores greater than the mean=75?
scores[scores>75]
```

```python
len(scores[scores>75]) # len() returns the length of an array or list
```

```python
# what is the actual mean of scores? In theory it should be 75
np.mean(scores)
```

```python
# what if you wanted to learn more about a function, eg, np.mean() function?
help(np.mean)
```

# Let's make some viz!

```python
# We will use matplotlib library:
import matplotlib.pyplot as plt # we only need pyplot sublibrary from matplotlib, give alias 'plt' to it (convention)
```

```python
# make a histogram of playingtime:
plt.figure(figsize=(6,3))
plt.hist(df.playingtime, bins=1000, edgecolor='black', color='tomato')
plt.xlim(-10,1000) # this specifies the range of the x axis, -10 gives a slight white space which looks better
plt.xlabel('playing time')
plt.ylabel('# of board games')
plt.savefig('histogramofplayingtime.png',dpi=300, bbox_inches='tight')
plt.show() # gets rid unnecessary text that gets printed
```
